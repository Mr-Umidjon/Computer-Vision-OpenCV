# Filterlash

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/Toshkent.png', 0)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img, cmap='gray')
plt.show()
```

# Funksiyalar:
* filter2D
* threshold

```python
kernel = np.array([
    [0, -1, -1],
    [1, -4, 1],
    [0, -1, 1]
])

filter_img = cv2.filter2D(img, -1, kernel)

plt.imshow(filter_img, cmap='gray')
plt.show()
```

```python
kernel = np.array([
    [0, -1, -1],
    [1, 4, -1],
    [0, -1, -1]
])

filter_img = cv2.filter2D(img, -1, kernel)

plt.imshow(filter_img, cmap='gray')
plt.show()
```

# Identity kernel

```python
identity = np.array([
    [0, 0, 0],
    [0, 1, 0],
    [0, 0, 0]
])

filter_img = cv2.filter2D(img, -1, identity)

plt.imshow(filter_img, cmap='gray')
plt.show()
```

# Edge detection

```python
edge = np.array([
    [-1, -1, -1],
    [-1, 8, -1],
    [-1, -1, -1]
])

filter_img = cv2.filter2D(img, -1, edge)

plt.imshow(filter_img, cmap='gray')
plt.show()
```

# Sharpen kernel

```python
sharpen = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

filter_img = cv2.filter2D(img, -1, sharpen)

plt.imshow(filter_img, cmap='gray')
plt.show()
```

# Edge kernel

```python
edge = np.array([
    [0, -1, 0],
    [-1, 4, -1],
    [0, -1, 0]
])

filter_img = cv2.filter2D(img, -1, edge)

plt.imshow(filter_img, cmap='gray')
plt.show()
```

# Sobel x kernel

```python
sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

filter_img = cv2.filter2D(img, -1, sobel_x)

plt.imshow(filter_img, cmap='gray')
plt.show()
```

# Sobel y kernel

```python
sobel_y = np.array([
    [-1, -2, 1],
    [0, 0, 0],
    [-1, 2, 1]
])

filter_img = cv2.filter2D(img, -1, sobel_y)

plt.imshow(filter_img, cmap='gray')
plt.show()
```

# Gaussian Blur Kernel

```python
box = np.ones((3, 3)) * 1 / 9.

filter_img = cv2.filter2D(img, -1, box)

plt.imshow(filter_img, cmap='gray')
plt.show()
```

```python
blur = cv2.GaussianBlur(img, (5, 5), 0)
plt.imshow(blur, cmap='gray')
plt.show()
```

```python
blur = cv2.GaussianBlur(img, (5, 5), 0)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')

plt.subplot(1, 2, 2)
plt.imshow(blur, cmap='gray')
plt.show()
```

# Threshold

```python
sobel_y = np.array([
    [-1, -2, 1],
    [0, 0, 0],
    [-1, 2, 1]
])

filter_img_y = cv2.filter2D(img, -1, sobel_y)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(filter_img_y, cmap='gray')

filter_img_blur = cv2.filter2D(blur, -1, sobel_y)

plt.subplot(1, 2, 2)
plt.imshow(filter_img_blur, cmap='gray')

plt.show()
```

```python
ret, img_thresh = cv2.threshold(filter_img_blur, 30, 255, cv2.THRESH_BINARY_INV)
plt.imshow(img_thresh, cmap='gray')
plt.show()
```
