```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_dtm = cv2.imread('images/DTM.jpg')
img_dtm = cv2.cvtColor(img_dtm, cv2.COLOR_BGR2RGB)

plt.imshow(img_dtm)
plt.show()
```

```python
img_ans = cv2.imread('images/DTM - Answer.jpg')
img_ans = cv2.cvtColor(img_ans, cv2.COLOR_BGR2RGB)

plt.imshow(img_ans)
plt.show()
```

```python
true_answer = ['A', "B", "A", "D", "C", 'E', 'A', 'B', 'A', 'D']
```

```python
sing1 = np.array([200, 0, 0])
sing2 = np.array([255, 100, 100])

mask = cv2.inRange(img_ans, sing1, sing2)

plt.imshow(mask, cmap='gray')
plt.show()
```

```python
img_dtm_copy = np.copy(img_dtm)

plt.imshow(img_dtm_copy)
plt.show()
```

```python
img_dtm_copy[mask == 0] = 0
plt.imshow(img_dtm_copy)
plt.show()
```

```python
answers = []
for i in range(175, 785, 65):
    answers.append(img_dtm_copy[i:i + 50, 170:505])
```

```python

variants = []
for i in range(10):
    for j in range(0,335,70):
        variants.append(answers[i][:,j:j+52])  # Matrix bo'yicha append
plt.imshow(variants[0],cmap="gray")
```

```python
len(answers), len(variants)
```

```python
for i in range(len(answers)):
    plt.subplot(len(answers), 1, i + 1)
    plt.imshow(answers[i])
plt.show()
```

```python
fig, axes = plt.subplots(10,5,figsize=(10,20))

k = 0
for i in range(10):
    for j in range(5):
        axes[i][j].imshow(variants[k],cmap="gray")
        k+=1
plt.show()
```
