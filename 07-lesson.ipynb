```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# Fast Fourier Transform

```python
img = cv2.imread('images/Toshkent.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
def fourier(img):
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    H = 20 * np.log(np.abs(fshift))
    return H
```

```python
norm_img = img / 255.
print(norm_img)
```

```python
plt.imshow(norm_img)
plt.show()
```

```python
norm_img = fourier(norm_img)
```

```python
plt.imshow(norm_img)
plt.show()
```
