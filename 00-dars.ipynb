# Homework

```python
img = cv2.imread('images/Toshkent.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
def zoom_in_image(imgage, n):
    r, c = imgage.shape[:2]
    mask = np.zeros((r*n, c*n, 3), dtype=np.uint8)

    for i in range(n):
        for j in range(n):
            mask[i::n, j::n] = imgage

    return mask
```

```python
img_x_n = zoom_in_image(img, 4)

print(img_x_n)
```

```python
print(img.shape)
print(img_x_n.shape)
```

```python
plt.imshow(img_x_n)
plt.show()
```

```python
def rectangle_to_image(img, p1, p2, color, thickness=1):
    x1, y1 = p1
    x2, y2 = p2
    color = np.array(color)
    img[x1:x1+thickness, x2:y2] = color
    img[y1-thickness:y1, x2:y2] = color

    img[x1:y1, x2-thickness:x2] = color
    img[x1:y1, y2-thickness:y2] = color

    return img
```

```python
img_res = rectangle_to_image(img_x_n, (0, 400), (1000, 1500), (255, 0, 255), 30)

plt.imshow(img_res)
plt.show()
```
