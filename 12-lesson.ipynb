# Kmeans algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
a1 = np.random.random((100, 2)) * 0.2 + 1
a2 = np.random.random((100, 2)) * 0.4 + 2
a3 = np.random.random((100, 2)) * 0.6 + 3

A = np.concatenate((a1, a2, a3), axis=0)

plt.plot(A[:, 0], A[:, 1], ".")
plt.show()
```

```python
A = np.float32(A)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 0.1)
k = 3

ret, label, center = cv2.kmeans(A, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
a1 = A[label.flatten()==0]
a2 = A[label.flatten()==1]
a3 = A[label.flatten()==2]

plt.scatter(a1[:, 0], a1[:, 1])
plt.scatter(a2[:, 0], a2[:, 1])
plt.scatter(a3[:, 0], a3[:, 1])
plt.scatter(center[:, 0], center[:, 1])
plt.show()
```

```python
img = cv2.imread('images/HSV_car.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
img_pixel = img.reshape((-1, 3))
img_pixel.shape
```

```python
img_pixel = np.float32(img_pixel)
criteria = (cv2.TERM_CRITERIA_MAX_ITER + cv2.TERM_CRITERIA_EPS, 10, 0.1)
k = 2

ret, label, center = cv2.kmeans(img_pixel, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
np.unique(label)
```

```python
label = label.reshape(img.shape[0], img.shape[1])
```

```python
center = np.uint8(center)
```

```python
k_means_img = center[label.flatten()]
```

```python
k_means_img.shape
```

```python
k_means_img = k_means_img.reshape(img.shape)

plt.imshow(k_means_img)
plt.show()
```
