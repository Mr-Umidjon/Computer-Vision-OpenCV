```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
video_green = cv2.VideoCapture('videos/uTorrent.mp4')
video_simple = cv2.VideoCapture('videos/imaginefor1minute.mp4')

output_video = cv2.VideoWriter("videos/output_video.avi", cv2.VideoWriter_fourcc(*"DIVX"), 30, (1920, 1080))

range_1 = np.array([200, 0, 0])
range_2 = np.array([255, 60, 50])

frames = []

while video_green.isOpened() and video_simple.isOpened():

    _, frame1 = video_green.read()
    __, frame2 = video_simple.read()

    if not (_ and __):
        break

    frame1 = cv2.resize(frame1, dsize=(1920, 1080))
    frame2 = cv2.resize(frame2, dsize=(1920, 1080))

    mask = cv2.inRange(frame1, range_1, range_2)

    frame1[mask != 0] = 0
    frame2[mask == 0] = 0

    frame3 = frame1 + frame2

    frames.append(frame3)

for frame in frames:
    output_video.write(frame)

video_simple.release()
video_green.release()
output_video.release()
cv2.destroyAllWindows()
```
